### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep = '|')
users.head(20)
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
    
users['gender_num'] = users['gender'].apply(gender_to_numeric)
users.head()
```

```python
temp = users.groupby('occupation').gender_num.sum() / users.occupation.value_counts()
temp.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
male = users.groupby('occupation').gender_num.sum()
female = users.occupation.value_counts() - users.groupby('occupation').gender_num.sum()
```

```python
columns = ['Male', 'Female']
c = pd.concat([male, female], axis=1)
c.columns = ['Male', 'Female']
c['Count'] = c.sum(axis=1)
c.head()
```

```python
def f1(x):
    return (x['Male']/x['Count'])*100
def f2(x):
    return (x['Female']/x['Count'])*100
    
c['Male'] = c.apply(f1, axis=1)
c['Female'] = c.apply(f2, axis=1)
c = c.drop('Count', axis = 1)
c.head(25)
```
